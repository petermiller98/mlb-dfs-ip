```python
import numpy as np
import pandas as pd
from pulp import *
import seaborn as sns
```

```python
# Read in the data
path = 'DK07312020.csv'
data = pd.read_csv(path)
```

```python
# rename/clean/drop columns
data = data.drop(columns=['Name + ID'])
data = data.rename(columns={'TeamAbbrev': 'Team'})
data = data.rename(columns={'AvgPointsPerGame': 'APPG'})
data = data.rename(columns={'Roster Position': 'RosterPosition'})
data = data.rename(columns={'Game Info': 'GameInfo'})
columns = ['Position', 'NameID', 'Name', 'ID', 'RosterPosition', 'GameInfo', 'Team']
data['GameInfo'] = data['GameInfo'].str.extract(r'([^\s]+)')
```

```python
# Add Opposing Team Column
data['OppTeam'] = data['GameInfo'].str.split('@')
data['OppTeam'] = [j[1] if i == j[0] else j[0] for i, j in zip(data['Team'], data['OppTeam'])]
```

```python
### Optional Filtering ###
removal  = ['Shane Bieber', 'Zach Plesac', 'Trevor Bauer', 'Sonny Gray', 'Carlos Carrasco', 
            'Merrill Kelly', 'Lance Lynn', 'Tyler Chatwood', 'Justin Verlander',
            'Aaron Civale', 'Eric Hosmer', 'Dylan Bundy', 'German Marquez', 'Gerrit Cole',
           'Tyler Glasnow', 'Jose Iglesias', 'Tyler Stephenson']
for name in removal:
    data = data[data['Name'] != name]

data = data.reset_index()
data = data.drop('index', axis=1)
data
```

```python
# One hot encoding of position
P = np.zeros(data.shape[0], dtype=int)
B1 = np.zeros(data.shape[0], dtype=int)
B2 = np.zeros(data.shape[0], dtype=int)
B3 = np.zeros(data.shape[0], dtype=int)
C = np.zeros(data.shape[0], dtype=int)
SS = np.zeros(data.shape[0], dtype=int)
OF = np.zeros(data.shape[0], dtype=int)

for i in range(data.shape[0]):
    if data.iloc[i, 0].find('SP') != -1:
        np.put(P, i, 1) 
    if data.iloc[i, 0].find('RP') != -1:
        np.put(P, i, 1)
    if data.iloc[i, 0].find('1B') != -1:
        np.put(B1, i, 1)
    if data.iloc[i, 0].find('2B') != -1:
        np.put(B2, i, 1)
    if data.iloc[i, 0].find('3B') != -1:
        np.put(B3, i, 1)
    if data.iloc[i, 0].find('C') != -1:
        np.put(C, i, 1)
    if data.iloc[i, 0].find('SS') != -1:
        np.put(SS, i, 1)
    if data.iloc[i, 0].find('OF') != -1:
        np.put(OF, i, 1)
data['P'] = P
data['1B'] = B1
data['2B'] = B2
data['3B'] = B3
data['C'] = C
data['SS'] = SS
data['OF'] = OF
```

```python
# number of teams playing 
teams = list(data['Team'].unique())
num_teams = len(teams)
num_players = data.shape[0]
```

```python
#one hot encoding team, game, and opposing team
one_hot = pd.get_dummies(data['Team'], prefix='Team')
data = data.join(one_hot)
one_hot2 = pd.get_dummies(data['GameInfo'])
data = data.drop('GameInfo', axis = 1)
data = data.join(one_hot2)
one_hot = pd.get_dummies(data['OppTeam'], prefix='Opp')
data = data.join(one_hot)

pd.set_option('display.max_columns', None)
data.head()
```

```python
# Dictionaries with index as key 
player = [str(i) for i in range(num_players)]
point = {str(i): data['APPG'][i] for i in range(num_players)} 
cost = {str(i): data['Salary'][i] for i in range(num_players)}
P = {str(i): data['P'][i] for i in range(num_players)}
B1 = {str(i): data['1B'][i] for i in range(num_players)}
B2 = {str(i): data['2B'][i] for i in range(num_players)}
B3 = {str(i): data['3B'][i] for i in range(num_players)}
C = {str(i): data['C'][i] for i in range(num_players)}
SS = {str(i): data['SS'][i] for i in range(num_players)}
OF = {str(i): data['OF'][i] for i in range(num_players)}
```

```python
#dicts = []
#for team in teams: 
#    dicts.append({str(i): data[team] for i in range(data.shape[0])})
#dicts[1]
```

```python
prob = LpProblem("DFS_Baseball", LpMaximize)
```

```python
player_vars = LpVariable.dicts("Players",player,0,1,LpBinary)
```

```python
#OBJECTIVE FUNCTION
prob += lpSum([point[i]*player_vars[i] for i in player]), "Points"
```

```python
#CONSTRAINTS
prob += lpSum([player_vars[i] for i in player]) == 10, "10_Players"
prob += lpSum([cost[i] * player_vars[i] for i in player]) <= 50000, "Total_Cost"
prob += lpSum([P[i] * player_vars[i] for i in player]) == 2, "2Pitchers"
prob += lpSum([B1[i] * player_vars[i] for i in player]) == 1 | 0, "1B"
prob += lpSum([B2[i] * player_vars[i] for i in player]) == 1 | 0, "2B"
prob += lpSum([B3[i] * player_vars[i] for i in player]) == 1 | 0, "3B"
prob += lpSum([C[i] * player_vars[i] for i in player]) == 1 | 0, "C"
prob += lpSum([SS[i] * player_vars[i] for i in player]) == 1 | 0, "SS"
prob += lpSum([OF[i] * player_vars[i] for i in player]) == 3, "OF"
```

```python
# at least 3 players on same team
num_teams = len(teams)
used_3_in_team = [LpVariable("u{}".format(i + 1), cat="Binary") for i in range(num_teams)]
i = 0
for team in teams:
    prob += (3 * used_3_in_team[i] <= 
             lpSum(data['Team_' + team][k] * player_vars.get(str(k)) for k in range(num_players)))
    i += 1
prob += pulp.lpSum(used_3_in_team) >= 1
```

```python
# no hitters vs pitcher
hitter_pitcher = [LpVariable("u{}".format(i + 1), cat="Binary") for i in range(num_teams)]
i = 0
for team1 in teams:
    for team2 in teams:
        prob += (8 * lpSum(P.get(str(k)) * player_vars.get(str(k)) * data['Team_' + team1][k] for k in range(num_players)) +
            lpSum((1 - P.get(str(k))) * player_vars.get(str(k)) * data['Opp_' + team2][k] for k in range(num_players)) <= 8)
```

```python
status = prob.solve()
```

```python
print("Status:", LpStatus[prob.status])
```

```python
selection = {}
for v in prob.variables():
    if len(v.name.split("_")) == 2:
        index = int(v.name.split("_")[1])
        selection[index] = v.varValue
```

```python
data['integer_programming'] = 0.0
for i in selection:
    data.loc[i, 'integer_programming'] = selection[i]
```

```python
lineup = data[data['integer_programming'] == 1.0]
total_points = lineup['APPG'].sum()
total_cost = lineup['Salary'].sum()
total_players = lineup.shape[0]
print("Total points: {}\nCost: ${}\nNumber of players: {}".format(total_points, total_cost, total_players))
```
