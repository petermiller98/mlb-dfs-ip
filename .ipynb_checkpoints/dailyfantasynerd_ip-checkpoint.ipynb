```python
import numpy as np
import pandas as pd
from pulp import *
import seaborn as sns
```

### Read in Data

```python
dfn_hitters = pd.read_csv('DFN_Hitters_08012020.csv')
dfn_hitters = dfn_hitters[['Player Name', 'Pos', 'Salary', 'Team', 'Opp', 'Proj FP']]
dfn_hitters = dfn_hitters.rename(columns={'Player Name': 'Name',
                                    'Pos' : 'Position'})
```

```python
dfn_pitchers = pd.read_csv('DFN_Pitchers_08012020.csv')
dfn_pitchers = dfn_pitchers[['Player Name', 'Pos', 'Salary', 'Team', 'Opp', 'Proj FP']]
dfn_pitchers = dfn_pitchers.rename(columns={'Player Name': 'Name',
                                    'Pos' : 'Position'})
```

```python
data = pd.concat([dfn_pitchers, dfn_hitters])
data['Opp'] = data['Opp'].str.replace('@', '')
data
```

### Filtering of non-starting pitchers and DTD or DL Hitters

```python
### Optional Filtering ###
removal  = []
for name in removal:
    data = data[data['Name'] != name]

data = data.reset_index()
data = data.drop('index', axis=1)
```

### One Hot Encoding of Position

```python
P = np.zeros(data.shape[0], dtype=int)
B1 = np.zeros(data.shape[0], dtype=int)
B2 = np.zeros(data.shape[0], dtype=int)
B3 = np.zeros(data.shape[0], dtype=int)
C = np.zeros(data.shape[0], dtype=int)
SS = np.zeros(data.shape[0], dtype=int)
OF = np.zeros(data.shape[0], dtype=int)

for i in range(data.shape[0]):
    if data.iloc[i, 1].find('SP') != -1:
        np.put(P, i, 1) 
    if data.iloc[i, 1].find('RP') != -1:
        np.put(P, i, 1)
    if data.iloc[i, 1].find('1B') != -1:
        np.put(B1, i, 1)
    if data.iloc[i, 1].find('2B') != -1:
        np.put(B2, i, 1)
    if data.iloc[i, 1].find('3B') != -1:
        np.put(B3, i, 1)
    if data.iloc[i, 1].find('C') != -1:
        np.put(C, i, 1)
    if data.iloc[i, 1].find('SS') != -1:
        np.put(SS, i, 1)
    if data.iloc[i, 1].find('OF') != -1:
        np.put(OF, i, 1)
data['P'] = P
data['1B'] = B1
data['2B'] = B2
data['3B'] = B3
data['C'] = C
data['SS'] = SS
data['OF'] = OF
```

```python
# number of teams playing 
teams = list(data['Team'].unique())
#games = list(data['GameInfo'].unique())
num_teams = len(teams)
num_players = data.shape[0]
#num_games = len(games)
```

### One Hot Encoding of Team, Opposing Team, and Game

```python
#one hot encoding team, game, and opposing team
one_hot = pd.get_dummies(data['Team'], prefix='Team')
data = data.join(one_hot)
#one_hot2 = pd.get_dummies(data['GameInfo'])
#data = data.drop('GameInfo', axis = 1)
#data = data.join(one_hot2)
one_hot3 = pd.get_dummies(data['Opp'], prefix='Opp')
data = data.join(one_hot3)
```

```python
def prepareArguments(data, old_lineups, num_players=data.shape[0]):
    """Returns list of dictionaries that are arguments for generateLineups function"""
    player = [str(i) for i in range(num_players)]
    point = {str(i): data['Proj FP'][i] for i in range(num_players)} 
    cost = {str(i): data['Salary'][i] for i in range(num_players)}
    P = {str(i): data['P'][i] for i in range(num_players)}
    B1 = {str(i): data['1B'][i] for i in range(num_players)}
    B2 = {str(i): data['2B'][i] for i in range(num_players)}
    B3 = {str(i): data['3B'][i] for i in range(num_players)}
    C = {str(i): data['C'][i] for i in range(num_players)}
    SS = {str(i): data['SS'][i] for i in range(num_players)}
    OF = {str(i): data['OF'][i] for i in range(num_players)}
    return [player, point, cost, P, B1, B2, B3, C, SS, OF, old_lineups]
```

```python
player = [str(i) for i in range(num_players)]
point = {str(i): data['Proj FP'][i] for i in range(num_players)} 
cost = {str(i): data['Salary'][i] for i in range(num_players)}
P = {str(i): data['P'][i] for i in range(num_players)}
B1 = {str(i): data['1B'][i] for i in range(num_players)}
B2 = {str(i): data['2B'][i] for i in range(num_players)}
B3 = {str(i): data['3B'][i] for i in range(num_players)}
C = {str(i): data['C'][i] for i in range(num_players)}
SS = {str(i): data['SS'][i] for i in range(num_players)}
OF = {str(i): data['OF'][i] for i in range(num_players)}
old_lineups = {str(i): 0 for i in range(num_players)}
```

### Instantiate Problem

```python
def generateLineup(data, player, point, cost, P, B1, B2, B3, C, SS, OF, old_lineups):
    """Returns list of names of optimal lineup given constraints."""
    prob = LpProblem("DFS_Baseball", LpMaximize)
    player_vars = LpVariable.dicts("Players",player,0,1,LpBinary)
    prob += lpSum([point[i]*player_vars[i] for i in player]), "Points"
    prob += lpSum([player_vars[i] for i in player]) == 10, "10_Players"
    prob += lpSum([cost[i] * player_vars[i] for i in player]) <= 50000, "Total_Cost"
    prob += lpSum([P[i] * player_vars[i] for i in player]) == 2, "2Pitchers"
    prob += lpSum([B1[i] * player_vars[i] for i in player]) == 1, "1B"
    prob += lpSum([B2[i] * player_vars[i] for i in player]) == 1, "2B"
    prob += lpSum([B3[i] * player_vars[i] for i in player]) == 1, "3B"
    prob += lpSum([C[i] * player_vars[i] for i in player]) == 1, "C"
    prob += lpSum([SS[i] * player_vars[i] for i in player]) == 1, "SS"
    prob += lpSum([OF[i] * player_vars[i] for i in player]) == 3, "OF"
    prob += lpSum([old_lineups.get(str(i)) * player_vars.get(str(i)) for i in range(num_players)]) <= 6, "Overlap"
    num_teams = len(teams)
    used_4_in_team = [LpVariable("u{}".format(i + 1), cat="Binary") for i in range(num_teams)]
    i = 0
    for team in teams:
        prob += (4 * used_4_in_team[i] <= 
             lpSum(data['Team_' + team][k] * player_vars.get(str(k)) for k in range(num_players)))
        i += 1
    prob += lpSum(used_4_in_team) >= 1
    for team1 in teams:
        prob += ((8 * lpSum(P.get(str(k)) * player_vars.get(str(k)) * data['Team_' + team1][k] for k in range(num_players)) +
            lpSum((1 - P.get(str(k))) * player_vars.get(str(k)) * data['Opp_' + team1][k] for k in range(num_players)) <= 8))
    status = prob.solve()
    #print("Status:", LpStatus[prob.status], "\n")
    selection = {}
    for v in prob.variables():
        if len(v.name.split("_")) == 2:
            index = int(v.name.split("_")[1])
            selection[index] = v.varValue
    data['integer_programming'] = 0.0
    for i in selection:
        data.loc[i, 'integer_programming'] = selection[i]
    lineup = data[data['integer_programming'] == 1.0]
    indexes = list(data.index[data['integer_programming'] == 1.0])
    for idx in indexes:
        old_lineups[str(idx)] = 1
    lineup = data[data['integer_programming'] == 1.0]
    total_points = lineup['Proj FP'].sum()
    total_cost = lineup['Salary'].sum()
    total_players = lineup.shape[0]
    print("Total points: {}\nCost: ${}\nNumber of players: {}".format(total_points, total_cost, total_players))
    return [lineup['Name'], old_lineups]
```

### Set at least 4 players on same team constraint
